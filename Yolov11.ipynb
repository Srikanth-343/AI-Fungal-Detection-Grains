```python
from ultralytics import YOLO

# Load the YOLO pretrained model
model = YOLO("yolov5s.pt")

# Train the model on a high-resolution dataset with enhanced augmentations
model.train(
    data="dataset.yaml",   # Path to your dataset config
    epochs=50,            # Train for more epochs
    imgsz=1280,            # High-resolution training
    batch=8,               # Adjust batch size based on memory
    lr0=0.01,              # Learning rate
    mosaic=True,           # Enable mosaic augmentation
    mixup=0.2,             # Apply mixup augmentation
    #rotation=10,           # Slight rotation augmentation
    #flip=True,             # Horizontal and vertical flips
    #blur=0.5,              # Apply Gaussian blur
    hsv_s=0.1,             # Slight color jitter
    conf=0.2,              # Fine-tune confidence threshold
    iou=0.6                # Adjust IoU for NMS
)
```
